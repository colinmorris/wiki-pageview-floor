```python
import matplotlib.patches as patches
from matplotlib import cm

dat = [
    (.15, 'List of longest fish'),
    (.2, 'Musca depicta'),
    (.35, 'Cow Tools'),
    (.81, 'Fox tossing'),
    (.96, 'List of presidents of the US with facial hair'),
]

FS = (7,5)
fig, ax = plt.subplots(figsize=FS)

def make_dartboard(ax, hilite=None, colorize=True):
    ax.set_ylim(1.0, 0.0)

    cmapname = 'Set2'
    cmap = cm.get_cmap(cmapname)

    prev = 0.0
    texts = []
    for i, (y, title) in enumerate(dat):
        anchor = (0, y)
        height = -(y - prev)
        width = .8
        color_ix = i / len(dat)
        color = cmap(color_ix) if colorize else 'white'
        rect = patches.Rectangle(anchor, width, height, color=color)
        #print(rect)
        ax.add_patch(rect)
        text_ynudge = -.0075
        text_xnudge = .005
        ax.plot([0, 0.1], [y, y], linewidth=2, color='black', alpha=.5)
        label = f'({y}) {title}'
        weight = 'normal'
        if title == hilite:
            weight='bold'
        text = ax.text(.1 + text_xnudge, y + text_ynudge, label,
                       fontweight=weight,
                      )
        texts.append(text)
        prev = y

    leftover = patches.Rectangle( (0, 1), .8, -.04,
                                 color=cmap(0) if colorize else 'white',
                                )
    ax.add_patch(leftover)
    
    ax.vlines(0.0, 0, 1)
    ax.set_facecolor( (1, 1, 1) )
    #ax.set_yticks([0.0, 1.0])
    ax.set_yticks([])
    label_xnudge = -.025
    ax.text(0 + label_xnudge, 0.02, '0')
    ax.text(0 + label_xnudge, 1.01, '1')
    #ax.set_ylabel('page_random')
    ax.text(-.09, -.05, 'page_random', family='monospace')
    ax.set_xticklabels([])

    
make_dartboard(ax, colorize=False)
ax.text(-.45, 1, ascii_archer2,
        family='monospace',
        fontweight='bold',
        fontsize=9,
       )

# TODO: Maybe this could just be an extra-image caption.
label_steps = 0
if label_steps:
    ax.text(-.6, .15,
            "1. Choose a random number from 0-1",
            fontsize=13,
            #fontweight='bold',
           )

fig.savefig('dartboard_1.png', dpi=144, bbox_inches='tight')
```
