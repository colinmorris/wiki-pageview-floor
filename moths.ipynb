```python
import pandas as pd
import seaborn as sns
sns.set_theme()
from matplotlib import pyplot as plt
import os

img_dir = "/home/colin/src/colinmorris.github.com/assets/wiki-floor/"
def savefig(fname, dpi=144):
    path = os.path.join(img_dir, fname)
    plt.savefig(path, bbox_inches='tight', dpi=dpi)


df = pd.read_csv('merged_phaeg.csv')
df.head()
```

```python
df['random_total'] = df.total.sample(frac=1, random_state=1).values
```

```python
df.head()
```

```python
FS = (8, 6)
fig, ax = plt.subplots(figsize=FS)
sns.scatterplot(data=df, x='gap', y='random_total', alpha=.5, marker='.')
ax.set_xlabel("Random gap")
ax.set_ylabel("2021 page views")
ax.set_ylim(0, 200)
ax.set_xlim(0, 
            #1e-6,
            7.5e-7,
           )
savefig("moth_gap_views_null_hypothesis.png")
```

```python
FS = (8, 6)
fig, ax = plt.subplots(figsize=FS)
sns.scatterplot(data=df, x='gap', y='total', alpha=.5, marker='.')
ax.set_xlabel("Random gap")
ax.set_ylabel("2021 page views")
ax.set_ylim(0, 200)
ax.set_xlim(0, 
            #1e-6,
            7.5e-7,
           )
savefig("moth_gap_views.png")
"""
ax.axline(xy1=(0, 3), xy2=(1.3265e-6, 256),
          color='red',
          dashes=(5,2),
         )
         """
```
